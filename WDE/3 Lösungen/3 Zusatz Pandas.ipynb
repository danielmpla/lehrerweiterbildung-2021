## Lösungshinweis:

Die Seite http://coronavirus.landkreis.gr/ verwendet auch für Unterseiten die selbe URL (der Webserver überschreibt die URL - sogenanntes URL rewriting). Um hier also die korrekte URL zu extrahieren, setzt
man auf die Fähigkeiten des Browsers, wobei sich 2 Möglichkeiten anbieten:

1) Auf der Hauptseite den Link zur Übersicht suchen und mit der rechten Maustaste anklicken und im aufklappenden Kontextmenü den Punkt "Link-Adresse kopieren" auswählen.

2) Alternativ kann man im Inspektor in den Entwicklertools das `frameset` inspizieren und findet dort die verwendeten Webadresse. Diese ermöglicht ihrerseits sich zu der Übersicht durchzuklicken, um dort, die korrekte URL bspw. aus der Adressleiste des Browsers zu kopieren.

Ergebnis in beiden Fällen sollte die folgende URL sein, die dann genutzt wird, um die Tabellendaten abzufragen:

https://www.kreis-goerlitz.de/Aktuelles/Coronavirus.htm/Seiten/Corona-Tabelle-Gemeinden.html?

## Lösung

```python
import pandas
import requests
#import lxml

response = requests.get(
    "https://www.kreis-goerlitz.de/Aktuelles/Coronavirus.htm/Seiten/Corona-Tabelle-Gemeinden.html?")

tables = pandas.read_html(response.text)

tables[0]
```

## Hinweis zum Ergebnis:

Die Ergebnistabelle enthält zum Schluss auch noch zwei unerwünschte Zeilen. Dies liegt daran, dass die Tabelle in der Webseite diese zwei zusätzlichen Zeilen zur Gestaltung verwendet, was man übrigens schön innerhalb des Inspektor's sehen kann. Dies ist zwar nicht unbedingt als "sauberes" Webdesign anzusehen, da es keine saubere Trennung zwischen Daten und Formattierung darstellt. Nichtsdestotrotz sind solche Formattierungen im Internet durchaus oft zu finden und stellen damit eine gängige Herausforderung für die Webdatenextraktion dar.
