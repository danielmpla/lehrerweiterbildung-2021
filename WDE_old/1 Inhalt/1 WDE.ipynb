## Webdatenextraktion vs. Crawling

Es gibt eine Reihe von Begriffen, die nicht immer ganz trennscharf zu unterscheiden sind. Man spricht hier bspw. von:
- Web Data Scraping und dem Web Crawling.

weitere Begrifflichkeiten:
- Scraper, Wrapper, Spider, Crawler, Bot, WebAgent, Grabber, Harvester, Web Trawler, Web Wanderer, Aggregator

Fragestellung: welche Informationen wollen wir?

- einzelne Daten einer Seite?
- alle Einträge (Records) einer Seite?
- Tabellendaten einer Seite?
- Informationen, die über mehrere Seiten verteilt sind? (bspw. über Detailseiten)
- alle Seiten einer Webseite?

### Wrapper

Zielstellung bei der Webdatenextraktion (WDE), ist es gewünschte Informationen aus Webdaten, sprich Webseiten oder direkt HTML-Seiten, mittels Wrappern zu extrahieren.

Wir konzentrieren uns dabei in diesem Kurs auf die technischen Grundlagen der manuellen Extraktion mit Hilfe von Python.

### Forschung

Wenn wir im Gegensatz dazu Forschungsfragen im Bereich WDE betrachten, ist dort meist die Zielstellung, den Prozess der Extraktion möglichst zu automatisieren und zu skalieren, um größere Mengen an Daten automatisch abzufragen.

## Der Browser: das wichtigste Tool

Die Entwicklertools der gängigen Browser (Firefox, Chrome, IE, Safari) bieten in der Funktionalität sehr vergleichbare Funktionalitäten, mit nur feinen unterschieden. Wir empfehlen für die Nutzung den Firefox.

### Entwicklertools aufrufen:

2 Möglichkeiten:

1) 'Alt'-Taste > Extras > Browser-Werkzeuge > Werkzeuge für Web-Entwickler
2) 'F12'-Taste

### Entwicklertools für WDE

Die 3 wichtigsten Tabs der Entwicklertools für WDE sind:
- Inspektor
- Konsole
- Netzwerkanalyse

Und um eine Seite zu inspizieren, die Möglichkeit, Elemente der Seite direkt auszuwählen:

### Inspektor

Der Inspektor ermöglicht es den geladenen DOM der Webseite zu inspizieren, so wie er sich nach dem Laden für den Browser darstellt. Durch einfaches Navigieren in dieser DOM-Ansicht mittels dem Mauszeiger, können Elemente gesucht und exploriert werden. Die Elemente über denen sich der Mauszeiger befindet werden dabei auf der Webseite gleich mit hervorgehoben, so dass eine Orientierung innerhalb der Webseite erleichtert wird.

### Konsole

Die Konsole bietet direkt im Kontext der Webseite, die Möglichkeit JavaScript auf den Daten der Webseite (u.a. dem DOM) auszuführen. Dies ist für WDE besonders hilfreich, um bspw. CSS Selektoren einfach zu testen.

Hierfür kann man bspw. in der Konsole die Abfrage mittels `$$(...)` nutzen.
Bspw. fragt `$$('div')` alle Div-Tags im DOM ab.

### Netzwerkanalyse

Bietet detaillierte Informationen, die in einer Webseite eingebetteten und geladenen Inhalte zu inspizieren, bspw. nach:
- Quelle
- Status
- Inhalt
- Dateityp
- Abfragemethode
- uvm.

## Was gilt es zu beachten

Wie mit fast jeder Technologie kann auch WDE sowohl für nützliche Zwecke aber genauso auch missbräuchlich verwendet werden. Hier gilt es einige Regeln zu beachten, damit es nicht zu etwaigen Konflikten kommt.

### Rechtliches

- Urheberrechte und Recht an der Datenbank berücksichtigen
    - Keine fremden Inhalte ins Netz stellen
    - keine kommerzielle Nutzung ohne expliziter Genehmigung
- AGBs berücksichtigen
    - wenn nicht explizit erlaubt: um Erlaubnis fragen

### Rücksichtsvolles Verhalten

Rücksichtsvolles Verhalten (eng. friendly behaviour) gegenüber Webseiteninhabern.

Heißt praktisch:
- Wartezeiten einbauen (kein "Hammering")
- die technischen Maßnahmen berücksichtigen (bspw. robots.txt)
- keine Schutzmaßnahmen umgehen
